## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

plt.figure(figsize=(8,6))
df['ConvertedComp'].plot(kind='box')
plt.show()
```

```python
plt.figure(figsize=(8,6))
df['Age'].plot(kind='box')
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
IQR = df['ConvertedComp'].quantile(0.75) - df['ConvertedComp'].quantile(0.25)
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here. finding lower bounds
LB = df['ConvertedComp'].quantile(0.25) - 1.5*IQR
LB = max(LB, df['ConvertedComp'].min())
LB
```

```python
# finding upper bounds
UB = df['ConvertedComp'].quantile(0.75) + 1.5*IQR
UB
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
len(df[df['ConvertedComp']>UB])
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df[df['ConvertedComp'] <= UB]
df2.head()
```

```python
# median of converted comp after removing outliers

df2['ConvertedComp'].median()
```

```python
# mean of converted comp after removing outliers

df2['ConvertedComp'].mean()
```
